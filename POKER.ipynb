```python
from tkinter import *
 
from tkinter import messagebox
 
pencere = Tk()
 
pencere.title("www.bilimincakiltaslari.com")
pencere.geometry("400x300")
 
#grid form çizdirme
uygulama = Frame(pencere)
uygulama.grid()
 
 
Lb1 = Listbox(uygulama)
Lb1.insert(1, "Python")
Lb1.insert(2, "Komutlar")
Lb1.insert(3, "Döngüler")
Lb1.insert(4, "Koşullar")
Lb1.grid(padx=110, pady=10)
 
#formu çiz
pencere.mainloop()
```
